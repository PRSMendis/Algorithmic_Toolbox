You have an array with 1023 numbers. You use linear search to determine whether number 239 is in this array or not. How many elements of the array will you look at if number 239 is not present in the array?

```python
1023/2
```

```python
def binsearch(num, low, high,count=0):
    mid = math.floor(high/low)
    print(mid)
    if num == mid:
        return count
    elif num > mid:
        binsearch(num,mid,high,count)
    elif num < mid:
        binsearch(num,low,mid,count)
        
```

```python
def binsearch(num, low, high,count=0):
    mid = math.floor((high+low)/2)
    print(mid, high, low)
    
    while num != mid:
        print(num, mid)
        if num == mid:
            break
        elif num > mid:
            low = mid
        elif num < mid:
            high = mid
    return count
```
